# Merging Datasets

```python
'''
Merging DBpedia and Spotify Datasets
'''
import pandas as pd
```

```python
#Reading Datasets and merging them
dbpediaData = pd.read_excel('dbpediaDataset.xlsx')
spotifyData = pd.read_excel('spotifyDataset.xlsx')

mergedData = spotifyData.merge(dbpediaData,on='Song')

mergedData.to_excel('CombinedData.xlsx',index=False)
```
